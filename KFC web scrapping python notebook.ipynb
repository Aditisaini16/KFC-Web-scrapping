```python
# importing required libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
source = requests.get("https://restaurants.kfc.co.in/location/delhi")
soup = BeautifulSoup(source.text, 'html.parser')
```

```python
name = soup.find_all('div', class_="store-info-box")
```

```python
store_name = []
for i in name:
    store_name.append(i.find('li', class_="outlet-name").text.strip())
```

```python
store_name
```

```python
address = []
for i in name:
    address.append(i.find('li', class_="outlet-address").text.strip())
```

```python
address
```

```python
phone_no = []
for i in name:
    phone_no.append(i.find('li', class_="outlet-phone").text.strip())
```

```python
phone_no
```

```python
timings = []
for i in name:
    timings.append((i.find('span', class_="sprite-icon intro-icon icn-time").next.next.next).text.strip())
```

```python
timings
```

```python
Coordinates = []
for x in name:
    Coordinates.append(x.find('input', class_="outlet-latitude")['value']+ ','+x.find('input', class_="outlet-longitude")['value'])
```

```python
Coordinates
```

```python
# for remaining 5 pages
for i in range(2,7):
    url = 'https://restaurants.kfc.co.in/location/delhi?page='+str(i)
    
    source = requests.get(url)
    soup = BeautifulSoup(source.text, 'html.parser')
    
    name = soup.find_all('div', class_="store-info-box")
    
    for i in name: 
        store_name.append(i.find('li', class_="outlet-name").text.strip())
        
    for i in name:
        address.append(i.find('li', class_="outlet-address").text.strip())
        
    for i in name:
        phone_no.append(i.find('li', class_="outlet-phone").text.strip())
        
    for i in name:
        timings.append((i.find('span', class_="sprite-icon intro-icon icn-time").next.next.next).text.strip())
        
    for x in name:
        Coordinates.append(x.find('input', class_="outlet-latitude")['value']+ ', '+
                           x.find('input', class_="outlet-longitude")['value'])
```

```python
lst = np.arange(1,61)
df = pd.DataFrame(list(zip(lst,store_name,address,phone_no,timings,Coordinates)), 
                    columns = ['S.No.','Store name','Address','Phone number','Timings','Coordinates'])
```

```python
df.set_index("S.No.", inplace=True)
df
```

```python
data = 'KFC Delhi stores data.xlsx'
df.to_excel(data)
```
